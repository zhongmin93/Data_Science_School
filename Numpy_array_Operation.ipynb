# 연습 문제 4.1.6
- 다음 배열은 첫번째 행(row)에 학번, 두번째 행에 영어 성적, 세번째 행에 수학 성적을 적은 배열이다. 영어 성적을 기준으로 각 열(column)을 재정렬하라.

array([[  1,    2,    3,    4],
       [ 46,   99,  100,   71],
       [ 81,   59,   90,  100]])

```python
h = np.array([[1,2,3,4],
             [46,99,100,71],
             [81,59,90,100]])
h
```

```python
# axis = 1은 행 기준, axis = 0은 열 기준
np.sort(h, axis=1)
```

```python
np.sort(h, axis=0) 
```
