# 정수 데이터 카운팅

```python
# 중복된 값 배제
np.unique([11, 11, 2, 2, 34, 34])
```

```python
a = np.array(['a','b','b','c','a'])
index, count = np.unique(a, return_counts=True)
```

```python
# 다중 할당
# a 에서 중복 제거한 값이 index에 할당되고
print('배열에 들어있는 원소 값 : ', index, '\n')

# a 에서 각 값을 가진 데이터 갯수가 count에 할당 된다.
print('a 개수 :',count[0]) 
print('b 개수 :',count[1]) 
print('c 개수 :',count[2]) 
```

- unique 명령은 데이터에 존재하는 값에 대해서만 갯수를 세므로, 배열 안에서 나올 수 있는 데이터이지만 나올 확률을 낮게 준다든가 등의 요인으로 인해 결과적으로 보여주지 않으면 count 되지 않는다.

- 예를 들어, 주사위 1~6까지 값 중에 6이 나오지 않았다면 6은 제외하고 count가 되어 5가 되므로 옳지 못하다.
